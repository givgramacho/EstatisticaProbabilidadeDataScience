#### Resumo de regressão 


- A **_análise de duas variavéis_** é determinada pela **_correlação entre elas_**, como e quanto essas **_variavéis quantitativas_** estão relacionandas.


- **_Regressão_** estuda determinadas variéval em função da outra, fornece um método de previsão que permite a avaliação em termos de confiabilidade.


- Permite a **_transformação de incerteza em risco_**.


- Analisar uma variavél dependente Y em relação a variável independente X é estudada em estatística por regressão, onde podemos determinar sua confiabilidade com métricas que determina coeficiente de Person ou **_coeficiente de correlação dado por "r"_** e **_coeficiente de determinação "$r^2$"_**.

### Exemplo prático de estudos de regressão 

- Para análisar o que está acontecendo podemos analisar uma representação de uma tabela em represenação gráfica.

- No gráfico temos podemos observar que, quanto maior a distância maior o tempo de entrega. 

- Quando aumenta a variável dependente tempo de entrega (em dia )no eixo Y aumenta a variável independente distância representado X. 

- Ao olhar para gráfico temos que fazer uma pergunta, existe uma relação entre essas variáveis? E a resposta é sim existe.



- E podemos imaginar uma reta que passa por esses pontos, isso significa que existe uma **_correlação entre essas variáveis_**.

- Podemos então calcular o coeficiente de correlação de Person "r" para ver a intensidade dessa correlação, se ela é uma correlação linear positiva, correlação linear negativa ou não existe correlação entre as variáveis. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Distância (Km)': [215, 480, 325, 550, 920, 670, 825, 1070, 1350, 1215],
                   'Tempo de entrega (dias)': [1, 1, 1.5, 2, 3, 3, 3.5, 4, 4.5, 5]})
print(df)
corr = df['Distância (Km)'].corr(df['Tempo de entrega (dias)'])
print("\n Correlação(%):",corr*100)

plt.scatter(df['Distância (Km)'], df['Tempo de entrega (dias)'])
plt.title("Tempo de entrega de carregamento de mercadoria por expedição rodoviária, em função da distância rodoviária percorrida ")
plt.xlabel('Distância Km(X)')
plt.ylabel('Tempo de entrega dias(Y)')
plt.show()
plt.grid()

import seaborn as sns

# Plotar o gráfico de dispersão com uma reta de regressão
sns.regplot(x='Distância (Km)', y='Tempo de entrega (dias)', data=df).set(title='Gráfico de Regressão Linear')
```

### Calculando o Coeficiente de Correlação "r" 

- Coeficiente de correlação determina o qual forte ou fraca e a relação dos dados em estudos. 




```python
import numpy as np

# Dados de exemplo
x = np.array([215, 480, 325, 550, 920, 670, 825, 1070, 1350, 1215])
y = np.array([1, 1, 1.5, 2, 3, 3, 3.5, 4, 4.5, 5])

# Calcular a média das variáveis x e y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Inicializar os valores de b e a como zero
b = 0
a = 0

# Calcular os valores de b e a
for i in range(len(x)):
    b += (x[i] - x_mean) * (y[i] - y_mean)
    a += (x[i] - x_mean)**2

b /= a
b *= 1.0
a = y_mean - b * x_mean

# Tempo para 900km

# Calcular o coeficiente de correlação (r)
r = np.sum((x - x_mean) * (y - y_mean)) / np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))

# Calcular o coeficiente de determinação (r^2)
r_squared = r**2


# Exibir os resultados
print("Equação da reta: y = {:.3f} + {:.4f}x".format(a, b))
print("Coeficiente de correlação (r): {:.2f}".format(r))
print("Coeficiente de determinação (r^2): {:.2f}".format(r_squared))
```

### Calculando o coeficiente de determinação "$r^2$".


- O coeficiente de determinação nos diz que 90% dos podem ser explicados, portanto os outros 10% não podem ser explicado pela equação. 

- Esses 10% também chamado de resíduos. 

### Código em python 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Dados para plotagem
x = np.array([215, 480, 325, 550, 920, 670, 825, 1070, 1350, 1215])
y = np.array([1, 1, 1.5, 2, 3, 3, 3.5, 4, 4.5, 5])

# Cálculo da equação da reta
a, b = np.polyfit(x, y, 1)

# Cálculo do coeficiente de correlação (r)
r, _ = pearsonr(x, y)

# Cálculo do coeficiente de determinação (r^2)
r2 = r**2

# Geração dos valores da reta para plotagem
line_x = np.linspace(x.min(), x.max(), 100)
line_y = a * line_x + b

# Criação do gráfico
fig, ax = plt.subplots()

# Adição dos dados ao gráfico
ax.plot(x, y, 'bo', label='Dados')

# Adição da reta ao gráfico
ax.plot(line_x, line_y, 'r-', label='Reta de ajuste')

# Adição da legenda
ax.legend(loc='best')

# Adição do título
ax.set_title('Gráfico com dados e reta de ajuste')

# Adição dos rótulos dos eixos
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')

# Adição da equação da reta e dos valores de r e r^2 ao gráfico
text = 'y = {:.2f}x + {:.2f}\n'.format(a, b)
text += 'r = {:.2f}\n'.format(r)
text += 'r^2 = {:.2f}'.format(r2)
ax.text(0.5, 0.5, text, ha='center', va='center', transform=ax.transAxes)

# Exibição do gráfico
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de serviço e Número de clientes
df = pd.DataFrame({'Nos': ['01', '08', '24', '38', '44', '58', '59', '72', '80', '92'],
                   'Matemática_xi': [5.0, 8.0, 7.0, 10.0, 6.0, 7.0, 9.0, 3.0, 8.0, 2.0],
                   'Estatística_yi': [6.0, 9.0, 8.0, 10.0, 5.0, 7.0, 8.0, 4.0, 6.0, 2.0]})
print(df)
# calcula a correlação entre Anos de serviço e Número de clientes
corr, p = pearsonr(df['Matemática_xi'], df['Estatística_yi'])
  
# imprime o coeficiente de correlação e o p-value
print(f'\n Coeficiente de correlação: {corr*100:.2f}% \n')
print(f'Coeficiente de determinação : {corr**2:.2f}\n')

# verifica o p-value e imprime a mensagem adequada
if corr == 1:
    print("Há uma correlação perfeita e positiva entre as variáveis")
elif (corr < 1 and corr>0.8):
    print("Correlação linear positiva altamente significativa entre as duas variáveis")

elif corr == -1:
    print("Há uma correlação perfeita a negativa entre as variáveis")
else:
    print('Não há correlação entre as variavéis ou a relação que porventura exista, não é linear')
    
# plota o gráfico de dispersão com a reta de regressão
sns.regplot(x='Matemática_xi', y='Estatística_yi', data=df).set(title=('Coeficiente de Person(r)'))


x1 = df['Matemática_xi']
y1 = df['Estatística_yi']
# Cálculo da equação da reta
a1, b1 = np.polyfit(x1, y1, 1)
# Cálculo do coeficiente de correlação (r)
r, _ = pearsonr(x1, y1)

# Cálculo do coeficiente de determinação (r^2)
r2 = r**2

# Geração dos valores da reta para plotagem
line_x = np.linspace(x1.min(), x1.max(), 100)
line_y = a1 * line_x + b1

# Criação do gráfico
fig, ax = plt.subplots()

# Adição dos dados ao gráfico
ax.plot(x1, y1, 'bo', label='Dados')

# Adição da reta ao gráfico
ax.plot(line_x, line_y, 'r-', label='Reta de ajuste')

# Adição da legenda
ax.legend(loc='best')

# Adição do título
ax.set_title('Gráfico com dados e reta de ajuste')

# Adição dos rótulos dos eixos
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')

# Adição da equação da reta e dos valores de r e r^2 ao gráfico
text = 'y = {:.2f}x + {:.2f}\n'.format(a1, b1)
text += 'r = {:.2f}\n'.format(r)
text += 'r^2 = {:.2f}'.format(r2)
ax.text(0.5, 0.5, text, ha='center', va='center', transform=ax.transAxes)

# Exibição do gráfico
plt.show()

# Exibir os resultados
print("Equação da reta: y = {:.2f} + {:.2f}x".format(a1, b1))
print("Coeficiente de correlação (r): {:.2f}".format(r))
print("Coeficiente de determinação (r^2): {:.2f}".format(r2))
```
