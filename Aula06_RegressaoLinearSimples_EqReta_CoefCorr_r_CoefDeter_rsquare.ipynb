#### Resumo de regresso 


- A **_anlise de duas variavis_**  determinada pela **_correlao entre elas_**, como e quanto essas **_variavis quantitativas_** esto relacionandas.


- **_Regresso_** estuda determinadas varival em funo da outra, fornece um mtodo de previso que permite a avaliao em termos de confiabilidade.


- Permite a **_transformao de incerteza em risco_**.


- Analisar uma variavl dependente Y em relao a varivel independente X  estudada em estatstica por regresso, onde podemos determinar sua confiabilidade com mtricas que determina coeficiente de Person ou **_coeficiente de correlao dado por "r"_** e **_coeficiente de determinao "$r^2$"_**.

### Exemplo prtico de estudos de regresso 

- Para anlisar o que est acontecendo podemos analisar uma representao de uma tabela em represenao grfica.

- No grfico temos podemos observar que, quanto maior a distncia maior o tempo de entrega. 

- Quando aumenta a varivel dependente tempo de entrega (em dia )no eixo Y aumenta a varivel independente distncia representado X. 

- Ao olhar para grfico temos que fazer uma pergunta, existe uma relao entre essas variveis? E a resposta  sim existe.



- E podemos imaginar uma reta que passa por esses pontos, isso significa que existe uma **_correlao entre essas variveis_**.

- Podemos ento calcular o coeficiente de correlao de Person "r" para ver a intensidade dessa correlao, se ela  uma correlao linear positiva, correlao linear negativa ou no existe correlao entre as variveis. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Distncia (Km)': [215, 480, 325, 550, 920, 670, 825, 1070, 1350, 1215],
                   'Tempo de entrega (dias)': [1, 1, 1.5, 2, 3, 3, 3.5, 4, 4.5, 5]})
print(df)
corr = df['Distncia (Km)'].corr(df['Tempo de entrega (dias)'])
print("\n Correlao(%):",corr*100)

plt.scatter(df['Distncia (Km)'], df['Tempo de entrega (dias)'])
plt.title("Tempo de entrega de carregamento de mercadoria por expedio rodoviria, em funo da distncia rodoviria percorrida ")
plt.xlabel('Distncia Km(X)')
plt.ylabel('Tempo de entrega dias(Y)')
plt.show()
plt.grid()

import seaborn as sns

# Plotar o grfico de disperso com uma reta de regresso
sns.regplot(x='Distncia (Km)', y='Tempo de entrega (dias)', data=df).set(title='Grfico de Regresso Linear')
```

### Calculando o Coeficiente de Correlao "r" 

- Coeficiente de correlao determina o qual forte ou fraca e a relao dos dados em estudos. 




```python
import numpy as np

# Dados de exemplo
x = np.array([215, 480, 325, 550, 920, 670, 825, 1070, 1350, 1215])
y = np.array([1, 1, 1.5, 2, 3, 3, 3.5, 4, 4.5, 5])

# Calcular a mdia das variveis x e y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Inicializar os valores de b e a como zero
b = 0
a = 0

# Calcular os valores de b e a
for i in range(len(x)):
    b += (x[i] - x_mean) * (y[i] - y_mean)
    a += (x[i] - x_mean)**2

b /= a
b *= 1.0
a = y_mean - b * x_mean

# Tempo para 900km

# Calcular o coeficiente de correlao (r)
r = np.sum((x - x_mean) * (y - y_mean)) / np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))

# Calcular o coeficiente de determinao (r^2)
r_squared = r**2


# Exibir os resultados
print("Equao da reta: y = {:.3f} + {:.4f}x".format(a, b))
print("Coeficiente de correlao (r): {:.2f}".format(r))
print("Coeficiente de determinao (r^2): {:.2f}".format(r_squared))
```

### Calculando o coeficiente de determinao "$r^2$".

### Cdigo em python 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Dados para plotagem
x = np.array([215, 480, 325, 550, 920, 670, 825, 1070, 1350, 1215])
y = np.array([1, 1, 1.5, 2, 3, 3, 3.5, 4, 4.5, 5])

# Clculo da equao da reta
a, b = np.polyfit(x, y, 1)

# Clculo do coeficiente de correlao (r)
r, _ = pearsonr(x, y)

# Clculo do coeficiente de determinao (r^2)
r2 = r**2

# Gerao dos valores da reta para plotagem
line_x = np.linspace(x.min(), x.max(), 100)
line_y = a * line_x + b

# Criao do grfico
fig, ax = plt.subplots()

# Adio dos dados ao grfico
ax.plot(x, y, 'bo', label='Dados')

# Adio da reta ao grfico
ax.plot(line_x, line_y, 'r-', label='Reta de ajuste')

# Adio da legenda
ax.legend(loc='best')

# Adio do ttulo
ax.set_title('Grfico com dados e reta de ajuste')

# Adio dos rtulos dos eixos
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')

# Adio da equao da reta e dos valores de r e r^2 ao grfico
text = 'y = {:.2f}x + {:.2f}\n'.format(a, b)
text += 'r = {:.2f}\n'.format(r)
text += 'r^2 = {:.2f}'.format(r2)
ax.text(0.5, 0.5, text, ha='center', va='center', transform=ax.transAxes)

# Exibio do grfico
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de servio e Nmero de clientes
df = pd.DataFrame({'Nos': ['01', '08', '24', '38', '44', '58', '59', '72', '80', '92'],
                   'Matemtica_xi': [5.0, 8.0, 7.0, 10.0, 6.0, 7.0, 9.0, 3.0, 8.0, 2.0],
                   'Estatstica_yi': [6.0, 9.0, 8.0, 10.0, 5.0, 7.0, 8.0, 4.0, 6.0, 2.0]})
print(df)
# calcula a correlao entre Anos de servio e Nmero de clientes
corr, p = pearsonr(df['Matemtica_xi'], df['Estatstica_yi'])
  
# imprime o coeficiente de correlao e o p-value
print(f'\n Coeficiente de correlao: {corr*100:.2f}% \n')
print(f'Coeficiente de determinao : {corr**2:.2f}\n')

# verifica o p-value e imprime a mensagem adequada
if corr == 1:
    print("H uma correlao perfeita e positiva entre as variveis")
elif (corr < 1 and corr>0.8):
    print("Correlao linear positiva altamente significativa entre as duas variveis")

elif corr == -1:
    print("H uma correlao perfeita a negativa entre as variveis")
else:
    print('No h correlao entre as variavis ou a relao que porventura exista, no  linear')
    
# plota o grfico de disperso com a reta de regresso
sns.regplot(x='Matemtica_xi', y='Estatstica_yi', data=df).set(title=('Coeficiente de Person(r)'))


x1 = df['Matemtica_xi']
y1 = df['Estatstica_yi']
# Clculo da equao da reta
a1, b1 = np.polyfit(x1, y1, 1)
# Clculo do coeficiente de correlao (r)
r, _ = pearsonr(x1, y1)

# Clculo do coeficiente de determinao (r^2)
r2 = r**2

# Gerao dos valores da reta para plotagem
line_x = np.linspace(x1.min(), x1.max(), 100)
line_y = a1 * line_x + b1

# Criao do grfico
fig, ax = plt.subplots()

# Adio dos dados ao grfico
ax.plot(x1, y1, 'bo', label='Dados')

# Adio da reta ao grfico
ax.plot(line_x, line_y, 'r-', label='Reta de ajuste')

# Adio da legenda
ax.legend(loc='best')

# Adio do ttulo
ax.set_title('Grfico com dados e reta de ajuste')

# Adio dos rtulos dos eixos
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')

# Adio da equao da reta e dos valores de r e r^2 ao grfico
text = 'y = {:.2f}x + {:.2f}\n'.format(a1, b1)
text += 'r = {:.2f}\n'.format(r)
text += 'r^2 = {:.2f}'.format(r2)
ax.text(0.5, 0.5, text, ha='center', va='center', transform=ax.transAxes)

# Exibio do grfico
plt.show()

# Exibir os resultados
print("Equao da reta: y = {:.2f} + {:.2f}x".format(a1, b1))
print("Coeficiente de correlao (r): {:.2f}".format(r))
print("Coeficiente de determinao (r^2): {:.2f}".format(r2))
```
