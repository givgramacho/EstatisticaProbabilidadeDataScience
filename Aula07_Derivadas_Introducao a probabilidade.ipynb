Visite [Aula 07 - Derivadas - Introduo  probabilidade Letscode](https://youtu.be/iydK5Gll6Fs) para mais informaes.

- Existe muitas situaes que envolvem incertezas: fenmenos ou experimentos aleatrios.

- Um modelo matemtico ajudar a investigar de maneira bastante precisa esse fenmeno.


- **_Encontramos na natureza dois tipos de fenmeno_**:

  - Determinsticos: Os resultados so sempre os mesmos e determinado pelas condies sob as quais o procedimento seja executado. 
  
  - No-Determinsticos: (Conhecido como Probabilsticos ou aleatrios) so aplicados em situaes que envolvem incertezas. Os resultados variam de uma observao a outra. 
  
  
- Fenmenos aleatrio, onde situaes ou acontecimentos cujos resultados no podem ser previstos com certeza, depende do acaso.
