### Aula 07 - Derivadas - Introdução à probabilidade

Visite [Aula 07 - Derivadas - Introdução à probabilidade Letscode](https://youtu.be/iydK5Gll6Fs) para mais informações.

- Existe muitas situações que envolvem incertezas: fenômenos ou experimentos aleatórios.

- Um modelo matemático ajudará a investigar de maneira bastante precisa esse fenômeno.


- **_Encontramos na natureza dois tipos de fenômeno_**:

  - Determinísticos: Os resultados são sempre os mesmos e determinado pelas condições sob as quais o procedimento seja executado. 
  
  - Não-Determinísticos: (Conhecido como Probabilísticos ou aleatórios) são aplicados em situações que envolvem incertezas. Os resultados variam de uma observação a outra. 
  
  
- Fenômenos aleatório, onde situações ou acontecimentos cujos resultados não podem ser previstos com certeza, depende do acaso.
