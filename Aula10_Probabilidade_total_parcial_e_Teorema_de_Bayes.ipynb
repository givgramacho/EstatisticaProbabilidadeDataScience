### Exemplo de pesquisa sobre um produto nessas diferentes cidades: 



- Probabilidade condicional de B dado A  definida:

$$\mathbb{P}(B|A) = \frac{\mathbb{P}(A \cap B )}{\mathbb{P}(A)}$$

a) Determine a probabilidade da pessoa ter dito sim, que conhece o produto

- ***Resp:***
$$\mathbb{P}(sim) = \frac{400}{1000} = 0,4 = 40\% $$


b) Determine a probabilidade da pessoa ser do rio de janeiro.

- ***Resp:***
$$\mathbb{P}(Ser RJ) = \frac{450}{1000} = 0,45 = 45\% $$

c) Determine a probabilidade da pessoa entrevistada ser de Belo Horizonte
- ***Resp:***
$$\mathbb{P}(Ser BH) = \frac{250}{1000} = 0,25 = 25\% $$

d) Determine a probabilidade da pessoa ter dito que no conhecia o produto dado que ela foi entrevistada em Belo Horizonte.

- ***Resp:***
$$\mathbb{P}(B|A) = \frac{\mathbb{P}(A \cap B )}{\mathbb{P}(A)}$$
$$\mathbb{P}(\text{no} |\text{BH}) = \frac{\mathbb{P}(\text{BH} \cap \text{no} )}{\mathbb{P}(\text{BH})}$$
$$\mathbb{P}(no|BH) = \frac{95}{250} = 0,38 = 38\% $$





### Teorema da probabilidade total



### Exemplo de Teorema da probabilidade total



$$\mathbb{P}(B) = \sum_{i=1}^{n}\mathbb{P}(B|A_i)\cdot\mathbb{P}(A_i)$$

$$\mathbb{P}(\text{spam}|\text{caixa1}) = 0.05 \rightarrow \mathbb{P}(\text{caixa1}) = 0.8$$

$$\mathbb{P}(\text{spam}|\text{caixa2}) = 0.10 \rightarrow \mathbb{P}(\text{caixa2}) = 0.2$$

$$\mathbb{P}(spam) = \sum_{i=1}^{n}\mathbb{P}(spam|caixa_i)\cdot\mathbb{P}(caixa_i)$$

$$\mathbb{P}(spam) = \mathbb{P}(spam|caixa_1)\cdot\mathbb{P}(caixa_1) + \mathbb{P}(spam|caixa_2)\cdot\mathbb{P}(caixa_2)$$

- Substituindo valores temos:

$$\mathbb{P}(spam) = 0.05\cdot 0.8 + 0.1\cdot 0.2 = 0.06 \rightarrow 6\%$$


### Teorema de Bayes



- Da probabilidade total conseguimos avanar e falar do teorema de Bayes. 


- Teorema de Bayes usa a probabilidade condicional usa tambm conceitos da probabilidade total para descrever que: Dado a probabilidade A, sendo que B ocorreu pela seguinte expresso 

$$\mathbb{P}(A|B) = \frac{\mathbb{P}(B|A)\cdot\mathbb{P}(A)}{\mathbb{P}(B)}$$



    -Probabilidade de ter a doena 
$\frac{1 \ \text{caso}}{1000 \ \text{habitantes}} \rightarrow \mathbb{P}(D) = 0.001 = 0.1\%$

        - Falso negativo
$\mathbb{P}(-|D) = 0.01 = 1\%$

        - Falso positivo
$\mathbb{P}(+|\bar{D}) = 0.05 = 5\%$


        - Probabilidade da Pessoa no ter a doena
$$\mathbb{P}(\bar{D}) = 1 - \mathbb{P}(D) $$
$\mathbb{P}(\bar{D}) = 1 - 0.1 = 0.999 $

##### Resolvendo o exerccio: 
$$\mathbb{P}(D|+) = \frac{\mathbb{P}(+|D)\cdot\mathbb{P}(D)}{\mathbb{P}(D)\cdot\mathbb{P}(+|D)+\mathbb{P}(\bar{D})\cdot\mathbb{P}(+|\bar{D})}$$

- Calculando e substituindo os valores

$\mathbb{P}(D|+) = \frac{0.99 x 0.001}{0.001x0.99 + 0.999x0.05} = 0.0194 = 1.94\%$, da probabilidade dessa pessoa ter doena dado que o resultado foi positivo.
