### Calcular e plotar Q1,Q2,Q3 e IQR 


```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a Pandas series with some sample data
data = pd.Series([1, 3, 5, 6, 8, 9, 11, 15, 16, 18, 20])

# Calculate the first, second, and third quartiles and the IQR
q1 = data.quantile(0.25)
q2 = data.quantile(0.50)
q3 = data.quantile(0.75)
iqr = q3 - q1

# Create the box plot
plt.boxplot([data.values], labels=['Data'], vert=True)
# Set the y-axis tick positions and labels
tick_positions = [q1, q2, q3, iqr]
tick_labels = [f'{q1:.2f}', f'{q2:.2f}', f'{q3:.2f}', f'{iqr:.2f}']
plt.yticks(tick_positions,tick_labels)

# Show the plot
plt.show()
print([q1,q2,q3,iqr])
```

### Variância e Desvio Padrão 

- Variância se baseia nos desvios em torno da média 

$s^2 = \frac{\sum{(x_i - \bar{x})^2}}{n - 1}$

$s^2$ = variância. 

$\bar{x}$ = média

- n = número de elementos da população 
- n - 1 = número de elementos da amostra 

- Desvio Padrão nada mais é que a raiz quadrada da variância. 

$s = \sqrt{\frac{\sum{(x_i - \bar{x})^2}}{n - 1}}$

#### Quanto maior o desvio padrão maior e a variação em torno da média



Calcule a mediana, média, moda, desvio padrão, variância, amplitude
total, primeiro quartil e terceiro quartil.

```python
import pandas as pd

data1 = pd.Series([10, 2 , 15, 11, 9, 3, 7, 15, 12, 5, 8])

sorted_data1 = data1.sort_values()

mediana = sorted_data1.median()
media = sorted_data1.mean()
moda = sorted_data1.mode()
desvio_padrao = sorted_data1.std()
variancia = sorted_data1.var()
amplitude_total = sorted_data1.max() - sorted_data1.min()
primeiro_quartil = round(sorted_data1.quantile(0.25))
terceiro_quartil = round(sorted_data1.quantile(0.75))

print([mediana,media,moda,desvio_padrao,variancia,amplitude_total, primeiro_quartil,terceiro_quartil])
print(sorted_data1)
sorted_data1.describe()
```
