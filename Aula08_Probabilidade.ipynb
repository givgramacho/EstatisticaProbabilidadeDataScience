### Definio de probabilidade 



### Exemplo de probabilidade *definio clssica*

- Ao lanar uma moeda temos a possibilidade de cair em cara e coroa 

- Ento dizemos que nosso 

$Espao Amostral(S) =  [cara(H),coroa(T)], n_{(S)} = 2$

- Evento (apenas obter cara)
$ A = [cara], n_{(A)} = 1$

- Portanto 

\begin{equation}\text{P(A)} = \frac{\text{n(A)}}{\text{n(S)}} = \frac{1}{2} = 0.50 = 50\% \end{equation}



### Exemplo de probabilidade *definio frequentista*

\begin{equation}f_n (A) = \frac{\text{n(A)}}{\text{n}}  \end{equation}

- fazendo a simulao abaixo vemos que chegamos ao valor clssico de 50% 


- Porque no usar a clssica? Porque neste contexto ela atende e outras problemas e necessrio usar a probabilidade frequntista que  mais eficiente. 



### Exemplo cdigo em python para probabilidade definio frequentista

- O cdigo Python para simular 10.000 lanamentos de moedas atribuindo um valor aleatrio de 0 ou 1 para cara e coroa . 


- Cada vez que a moeda  lanada, a probabilidade de obter cara ou coroa  de 50%, ento voc deve esperar que aproximadamente metade dos resultados sejam caras e metade sejam coroas (no ser exatamente a metade, devido a uma pequena variao aleatria ; mas deve ser prximo):

```python
%matplotlib inline
import random

# Create a list with 2 element (for heads and tails)
caras_coroas = [0,0]

# loop through 10000 trials
tentativas = 10000
tentativa = 0
while tentativa < tentativas:
    tentativa = tentativa + 1
    # Get a random 0 or 1
    lance = random.randint(0,1)
    # Increment the list element corresponding to the toss result
    caras_coroas[lance] = caras_coroas[lance] + 1

print (caras_coroas)

# Show a pie chart of the results
from matplotlib import pyplot as plt
plt.figure(figsize=(5,5))
plt.pie(caras_coroas, labels=['cara', 'coroa'])
plt.legend()
plt.show()
```

### Tipos de eventos 

### Probabilidade condicional e dependncia 

Os eventos podem ser:

- Independente (eventos que no so afetados por outros eventos)


- Dependente (eventos que so condicionais a outros eventos)


- Mutuamente Exclusivos (eventos que no podem ocorrer juntos)


### Exemplo de eventos independente

- Independente (eventos que no so afetados por outros eventos)


- Exemplo abaixo usando dois eventos independentes um evento no dependente do outro, ento **no** vamos ter uma **interseco**. 

$$\mathbb{P}(A \cap B ) = \mathbb{P}(A) * \mathbb{P}(B)$$



### Exemplo de eventos mutuamente exclusivos

- Mutuamente Exclusivos (eventos que no podem ocorrer juntos).


- Dizemos que os eventos so excludentes ou disjuntos quando a realizao de um exclui ou afeta a realizao do(s) outro(s).


- Assim, no lanamento de uma moeda,se no evento "tira cara", ento no  possvel "tirar coroa" se tirar um deles outro no se realizar. 


- Portanto chamamos de eventos excludentes, e eventos so mutuamente excludentes quando um exclui o outro.

Portanto 
$$(A \cap B ) =  \emptyset $$ 
$$\mathbb{P}(A \cap B ) = 0$$



### Probabilidade condicional 



- Probabilidade condicional de B dado A  definida:

$$\mathbb{P}(B|A) = \frac{\mathbb{P}(A \cap B )}{\mathbb{P}(A)}$$

-  utilizada quando quer saber a probabilidade de um evento dado a probabilidade de ocorrer um novo evento.
