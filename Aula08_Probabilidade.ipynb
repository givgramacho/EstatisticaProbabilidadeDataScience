### Definição de probabilidade 



### Exemplo de probabilidade *definição clássica*

- Ao lançar uma moeda temos a possibilidade de cair em cara e coroa 

- Então dizemos que nosso 

$Espaço Amostral(S) =  [cara(H),coroa(T)], n_{(S)} = 2$

- Evento (apenas obter cara)
$ A = [cara], n_{(A)} = 1$

- Portanto 

\begin{equation}\text{P(A)} = \frac{\text{n(A)}}{\text{n(S)}} = \frac{1}{2} = 0.50 = 50\% \end{equation}



### Exemplo de probabilidade *definição frequentista*

\begin{equation}f_n (A) = \frac{\text{n(A)}}{\text{n}}  \end{equation}

- fazendo a simulação abaixo vemos que chegamos ao valor clássico de 50% 


- Porque não usar a clássica? Porque neste contexto ela atende e outras problemas e necessário usar a probabilidade frequêntista que é mais eficiente. 



### Exemplo código em python para probabilidade definição frequentista

- O código Python para simular 10.000 lançamentos de moedas atribuindo um valor aleatório de 0 ou 1 para cara e coroa . 


- Cada vez que a moeda é lançada, a probabilidade de obter cara ou coroa é de 50%, então você deve esperar que aproximadamente metade dos resultados sejam caras e metade sejam coroas (não será exatamente a metade, devido a uma pequena variação aleatória ; mas deve ser próximo):

```python
%matplotlib inline
import random

# Create a list with 2 element (for heads and tails)
caras_coroas = [0,0]

# loop through 10000 trials
tentativas = 10000
tentativa = 0
while tentativa < tentativas:
    tentativa = tentativa + 1
    # Get a random 0 or 1
    lance = random.randint(0,1)
    # Increment the list element corresponding to the toss result
    caras_coroas[lance] = caras_coroas[lance] + 1

print (caras_coroas)

# Show a pie chart of the results
from matplotlib import pyplot as plt
plt.figure(figsize=(5,5))
plt.pie(caras_coroas, labels=['cara', 'coroa'])
plt.legend()
plt.show()
```

### Tipos de eventos 

### Probabilidade condicional e dependência 

Os eventos podem ser:

- Independente (eventos que não são afetados por outros eventos)


- Dependente (eventos que são condicionais a outros eventos)


- Mutuamente Exclusivos (eventos que não podem ocorrer juntos)


### Exemplo de eventos independente

- Independente (eventos que não são afetados por outros eventos)


- Exemplo abaixo usando dois eventos independentes um evento não dependente do outro, então **não** vamos ter uma **intersecção**. 

$$\mathbb{P}(A \cap B ) = \mathbb{P}(A) * \mathbb{P}(B)$$



### Exemplo de eventos mutuamente exclusivos

- Mutuamente Exclusivos (eventos que não podem ocorrer juntos).


- Dizemos que os eventos são excludentes ou disjuntos quando a realização de um exclui ou afeta a realização do(s) outro(s).


- Assim, no lançamento de uma moeda,se no evento "tira cara", então não é possível "tirar coroa" se tirar um deles outro não se realizará. 


- Portanto chamamos de eventos excludentes, e eventos são mutuamente excludentes quando um exclui o outro.

Portanto 
$$(A \cap B ) =  \emptyset $$ 
$$\mathbb{P}(A \cap B ) = 0$$



### Probabilidade condicional 



- Probabilidade condicional de B dado A é definida:

$$\mathbb{P}(B|A) = \frac{\mathbb{P}(A \cap B )}{\mathbb{P}(A)}$$

- É utilizada quando quer saber a probabilidade de um evento dado a probabilidade de ocorrer um novo evento.
