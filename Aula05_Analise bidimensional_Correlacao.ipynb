```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Agente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
                   'Anos de servio': [2, 3, 4, 5, 4, 6, 7, 8, 8, 10],
                   'Nmero de clientes': [48, 50, 56, 52, 43, 60, 62, 58, 64, 72]})
print(df)
corr = df['Anos de servio'].corr(df['Nmero de clientes'])
print("\n Correlao(%):",corr*100)

plt.scatter(df['Anos de servio'], df['Nmero de clientes'])
plt.title("Grfico de correlaao: Anos de servio e nmero de clientes")
plt.xlabel('Anos de servio(X)')
plt.ylabel('Nmero de clientes(Y)')
plt.grid()
plt.show()

import seaborn as sns

# Plotar o grfico de disperso com uma reta de regresso
sns.regplot(x='Anos de servio', y='Nmero de clientes', data=df).set(title='Grfico de Regresso Linear')
```

- Analisando o grfico acima, existe uma correlao linear positiva, quanto maior o anos de servios maior nmero de cliente.

### Coeficiente de correlao de Pearson (r) ou correlao linear 




```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de servio e Nmero de clientes
df = pd.DataFrame({'Agente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
                   'Anos de servio': [2, 3, 4, 5, 4, 6, 7, 8, 8, 10],
                   'Nmero de clientes': [48, 50, 56, 52, 43, 60, 62, 58, 64, 72]})

# calcula a correlao entre Anos de servio e Nmero de clientes
corr, p = pearsonr(df['Anos de servio'], df['Nmero de clientes'])
  
# imprime o coeficiente de correlao e o p-value
print(f'Coeficiente de correlao: {corr:.2f}')
print(f'p-value: {p:.2f}')

# verifica o p-value e imprime a mensagem adequada
if p == 1:
    print("H uma correlao perfeita e positiva entre as variveis")
elif p == -1:
    print("H uma correlao perfeita a negativa entre as variveis")
else:
    print('No h correlao entre as variavis ou a relao que porventura exista, no  linear')


# plota o grfico de disperso com a reta de regresso
sns.regplot(x='Anos de servio', y='Nmero de clientes', data=df).set(title=('Coeficiente de Person(r)'))
```

### Clculo de r de Person 



```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de servio e Nmero de clientes
df = pd.DataFrame({'Agente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
                   'Anos de servio': [2, 3, 4, 5, 4, 6, 7, 8, 8, 10],
                   'Nmero de clientes': [48, 50, 56, 52, 43, 60, 62, 58, 64, 72]})

# adiciona a coluna resultado
df = df.assign(XY=df['Anos de servio']*df['Nmero de clientes'])
df = df.assign(X2=df['Anos de servio']**2)
df = df.assign(Y2=df['Nmero de clientes']**2)



# calcula a correlao entre Anos de servio e Nmero de clientes
corr, p = pearsonr(df['Anos de servio'], df['Nmero de clientes'])
  

# plota o grfico de disperso com a reta de regresso
sns.regplot(x='Anos de servio', y='Nmero de clientes', data=df).set(title=('Coeficiente de Person(r)'))


# visualiza as primeiras linhas do dataframe


# imprime o coeficiente de correlao e o p-value
print(f'\nCoeficiente de correlao: {corr*100:.2f}% \n')
print(f'p-value: {p:.4f}\n')

# verifica o p-value e imprime a mensagem adequada
if p == 1:
    print("H uma correlao perfeita e positiva entre as variveis \n")
elif p == -1:
    print("H uma correlao perfeita a negativa entre as variveis \n")
else:
    print('No h correlao entre as variavis ou a relao que porventura exista, no  linear \n')

    
# Criar n=10
df.loc[10, 'Agente'] = 'n = 10'

# calcula a soma dos valores da coluna Anos de servio
soma = df['Anos de servio'].sum()
df.loc[10, 'Anos de servio'] = soma

# calcula a soma dos valores da coluna Nmero de clientes
soma = df['Nmero de clientes'].sum()
df.loc[10, 'Nmero de clientes'] = soma

# calcula a soma dos valores da coluna XY
soma = df['XY'].sum()
df.loc[10, 'XY'] = soma

# calcula a soma dos valores da coluna x2
soma = df['X2'].sum()
df.loc[10, 'X2'] = soma

# calcula a soma dos valores da coluna Y2
soma = df['Y2'].sum()
df.loc[10, 'Y2'] = soma

# visualiza o dataframe
df
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de servio e Nmero de clientes
df = pd.DataFrame({'Nos': ['01', '08', '24', '38', '44', '58', '59', '72', '80', '92'],
                   'Matemtica_xi': [5.0, 8.0, 7.0, 10.0, 6.0, 7.0, 9.0, 3.0, 8.0, 2.0],
                   'Estatstica_yi': [6.0, 9.0, 8.0, 10.0, 5.0, 7.0, 8.0, 4.0, 6.0, 2.0]})

# calcula a correlao entre Anos de servio e Nmero de clientes
corr, p = pearsonr(df['Matemtica_xi'], df['Estatstica_yi'])
  
# imprime o coeficiente de correlao e o p-value
print(f'\n Coeficiente de correlao: {corr*100:.2f}% \n')
print(f'p-value: {p:.2f}\n')

# verifica o p-value e imprime a mensagem adequada
if corr == 1:
    print("H uma correlao perfeita e positiva entre as variveis")
elif (corr < 1 or corr>0.8):
    print("Correlao linear positiva altamente significativa entre as duas variveis")
elif corr == -1:
    print("H uma correlao perfeita a negativa entre as variveis")
else:
    print('No h correlao entre as variavis ou a relao que porventura exista, no  linear')


# plota o grfico de disperso com a reta de regresso
sns.regplot(x='Matemtica_xi', y='Estatstica_yi', data=df).set(title=('Coeficiente de Person(r)'))
```
