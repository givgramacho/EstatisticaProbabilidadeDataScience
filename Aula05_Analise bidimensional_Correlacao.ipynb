- Quando temos uma variável usamos medida de tendência central ou variabilidade quando queremos resumir os dados, como vimos na aula anterior. 

- Quando temos mais de uma variável temos que usar análise bidimensional para explorar relações(similaridade) entre as variáveis 

- Quando temos duas **_variavéis quantitativas_**, a relação entre elas e chamada de **_correlação_**. 

- Exemplo: 



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Agente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
                   'Anos de serviço': [2, 3, 4, 5, 4, 6, 7, 8, 8, 10],
                   'Número de clientes': [48, 50, 56, 52, 43, 60, 62, 58, 64, 72]})
print(df)
corr = df['Anos de serviço'].corr(df['Número de clientes'])
print("\n Correlação(%):",corr*100)

plt.scatter(df['Anos de serviço'], df['Número de clientes'])
plt.title("Gráfico de correlaçao: Anos de serviço e número de clientes")
plt.xlabel('Anos de serviço(X)')
plt.ylabel('Número de clientes(Y)')
plt.grid()
plt.show()

import seaborn as sns

# Plotar o gráfico de dispersão com uma reta de regressão
sns.regplot(x='Anos de serviço', y='Número de clientes', data=df).set(title='Gráfico de Regressão Linear')
```

- Analisando o gráfico acima, existe uma correlação linear positiva, quanto maior o anos de serviços maior número de cliente. 

- Analisando o gráfico abaixo **_quanto maior a renda bruta mensal_**, **menor** é a **porcentagem da renda gasta em saúde**.Existe uma  Correlação linear negativa.

### Coeficiente de correlação de Pearson (r) ou correlação linear 




```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de serviço e Número de clientes
df = pd.DataFrame({'Agente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
                   'Anos de serviço': [2, 3, 4, 5, 4, 6, 7, 8, 8, 10],
                   'Número de clientes': [48, 50, 56, 52, 43, 60, 62, 58, 64, 72]})

# calcula a correlação entre Anos de serviço e Número de clientes
corr, p = pearsonr(df['Anos de serviço'], df['Número de clientes'])
  
# imprime o coeficiente de correlação e o p-value
print(f'Coeficiente de correlação: {corr:.2f}')
print(f'p-value: {p:.2f}')

# verifica o p-value e imprime a mensagem adequada
if p == 1:
    print("Há uma correlação perfeita e positiva entre as variáveis")
elif p == -1:
    print("Há uma correlação perfeita a negativa entre as variáveis")
else:
    print('Não há correlação entre as variavéis ou a relação que porventura exista, não é linear')


# plota o gráfico de dispersão com a reta de regressão
sns.regplot(x='Anos de serviço', y='Número de clientes', data=df).set(title=('Coeficiente de Person(r)'))
```

### Cálculo de r de Person 



```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de serviço e Número de clientes
df = pd.DataFrame({'Agente': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
                   'Anos de serviço': [2, 3, 4, 5, 4, 6, 7, 8, 8, 10],
                   'Número de clientes': [48, 50, 56, 52, 43, 60, 62, 58, 64, 72]})

# adiciona a coluna resultado
df = df.assign(XY=df['Anos de serviço']*df['Número de clientes'])
df = df.assign(X2=df['Anos de serviço']**2)
df = df.assign(Y2=df['Número de clientes']**2)



# calcula a correlação entre Anos de serviço e Número de clientes
corr, p = pearsonr(df['Anos de serviço'], df['Número de clientes'])
  

# plota o gráfico de dispersão com a reta de regressão
sns.regplot(x='Anos de serviço', y='Número de clientes', data=df).set(title=('Coeficiente de Person(r)'))


# visualiza as primeiras linhas do dataframe


# imprime o coeficiente de correlação e o p-value
print(f'\nCoeficiente de correlação: {corr*100:.2f}% \n')
print(f'p-value: {p:.4f}\n')

# verifica o p-value e imprime a mensagem adequada
if p == 1:
    print("Há uma correlação perfeita e positiva entre as variáveis \n")
elif p == -1:
    print("Há uma correlação perfeita a negativa entre as variáveis \n")
else:
    print('Não há correlação entre as variavéis ou a relação que porventura exista, não é linear \n')

    
# Criar n=10
df.loc[10, 'Agente'] = 'n = 10'

# calcula a soma dos valores da coluna Anos de serviço
soma = df['Anos de serviço'].sum()
df.loc[10, 'Anos de serviço'] = soma

# calcula a soma dos valores da coluna Número de clientes
soma = df['Número de clientes'].sum()
df.loc[10, 'Número de clientes'] = soma

# calcula a soma dos valores da coluna XY
soma = df['XY'].sum()
df.loc[10, 'XY'] = soma

# calcula a soma dos valores da coluna x2
soma = df['X2'].sum()
df.loc[10, 'X2'] = soma

# calcula a soma dos valores da coluna Y2
soma = df['Y2'].sum()
df.loc[10, 'Y2'] = soma

# visualiza o dataframe
df
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns

# cria o dataframe com as colunas Agente, Anos de serviço e Número de clientes
df = pd.DataFrame({'Nos': ['01', '08', '24', '38', '44', '58', '59', '72', '80', '92'],
                   'Matemática_xi': [5.0, 8.0, 7.0, 10.0, 6.0, 7.0, 9.0, 3.0, 8.0, 2.0],
                   'Estatística_yi': [6.0, 9.0, 8.0, 10.0, 5.0, 7.0, 8.0, 4.0, 6.0, 2.0]})

# calcula a correlação entre Anos de serviço e Número de clientes
corr, p = pearsonr(df['Matemática_xi'], df['Estatística_yi'])
  
# imprime o coeficiente de correlação e o p-value
print(f'\n Coeficiente de correlação: {corr*100:.2f}% \n')
print(f'p-value: {p:.2f}\n')

# verifica o p-value e imprime a mensagem adequada
if corr == 1:
    print("Há uma correlação perfeita e positiva entre as variáveis")
elif (corr < 1 or corr>0.8):
    print("Correlação linear positiva altamente significativa entre as duas variáveis")
elif corr == -1:
    print("Há uma correlação perfeita a negativa entre as variáveis")
else:
    print('Não há correlação entre as variavéis ou a relação que porventura exista, não é linear')


# plota o gráfico de dispersão com a reta de regressão
sns.regplot(x='Matemática_xi', y='Estatística_yi', data=df).set(title=('Coeficiente de Person(r)'))
```

### Resposta:

- Observe que o **valor-p é muito baixo**, indicando que a **correlação é muito provavelmente não devido ao acaso**. O **coeficiente de correlação é r= 91.12%**, o que indica uma **correlação linear positiva altamente significativa** entre as duas variáveis. Quem entende matemática também entende estatística.
